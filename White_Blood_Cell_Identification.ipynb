# **White Blood Cell Identification Project**
---
The "White Blood Cell Identification Project" is a research endeavor that specifically focuses on implementing the automated identification of white blood cells using deep learning methods on FPGA (Field Programmable Gate Array) platforms. By leveraging the power of FPGA technology, the project aims to optimize the performance and efficiency of the identification process. The dataset of white blood cell images is loaded and explored, and various data visualization techniques are utilized to analyze the distribution and statistics of different cell types. The ultimate goal is to develop an FPGA-based system that can accurately and efficiently classify neutrophils, eosinophils, basophils, lymphocytes, and monocytes in real-time scenarios.

```python

# Directory path containing the images
directory = 'TestA/'

### TODO ###
# import os
#folder = os.path.join('/content/drive/My Drive//content/drive/MyDrive/tubitak-2209-wbc-data/TestA')


data_list = []
# Iterate over the image files in the directory
for filename in os.listdir(directory):
    for image in os.listdir(directory+filename):
      if image.endswith(".jpg"):
        # Extract the image name, label, and label type from the filename
          image_name = image
          label = filename
        # Append the image name and label to the DataFrame
          data_list.append({"Image": image_name, "Label": label})
```

```python

df= pd.DataFrame(data_list)


```

```python

if not df.empty:
    label_counts = df["Label"].value_counts()
    plt.figure(figsize=(8, 6))
    label_counts.plot(kind='bar')
    plt.title('Label Distribution')
    plt.xlabel('Labels')
    plt.ylabel('Count')
    plt.show()


else:
    print("The DataFrame is empty.")
```
